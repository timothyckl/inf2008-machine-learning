## 1. Gini Impurity Formula

The Gini impurity of a dataset $(S)$ is defined as:

$$
\text{Gini}(S) = 1 - \sum_{k=1}^{K} p_k^2
$$

where $(p_k)$ is the proportion of samples in class $(k)$.


For a split of dataset \( S \) into subsets \( S_1, S_2, \dots, S_m \), the weighted Gini impurity is:

$$
\text{Gini}_{\text{split}} = \sum_{j=1}^{m} \frac{|S_j|}{|S|} \, \text{Gini}(S_j)
$$

The best split is the one that minimizes $\text{Gini}_{\text{split}}$.

2. At the current node, evaluate each candidate feature by:
   - Splitting the data into Yes/No groups
   - Computing the Gini impurity for each group
   - Computing the weighted Gini for the split
3. Select the feature with the lowest weighted Gini as the split.
4. Repeat Steps 2–3 for each resulting child node using the remaining features.
5. Stop splitting when the node is pure, no Gini reduction is achieved, or depth 2 is reached.

## 5. Reflection Questions

1. Why did Gini and Entropy choose different root splits in this dataset?

Gini and entropy measure impurity differently. Entropy penalizes class uncertainty more strongly near pure nodes, so it strongly favors splits that create a very pure child, even if the other child remains impure. Gini is less sensitive near purity and instead prefers splits that reduce impurity more evenly across both child nodes. In this dataset, one split created a perfectly pure node, which entropy rewarded more than Gini.

2. Does the fact that entropy produced a purer node mean it is the “better” criterion? Why or why not?

No. A purer node does not necessarily imply a better tree. Entropy and Gini reflect different design preferences, and neither guarantees better generalization. A split that produces a very pure node may overfit small or noisy datasets, while a more balanced split may generalize better. The choice of criterion is a modeling decision, not a measure of correctness.

3. In practice, would you expect Gini and Entropy to disagree often? Explain.

No. In most real-world datasets, Gini and entropy tend to rank candidate splits similarly, so they often select the same split. Disagreements are more likely in small or carefully constructed datasets, or when a split creates an extremely pure child node.
