## Part 2: Loss calculation

Recall the regression model:

$$
\hat{y} = wx + b

$$
L(w,b) = \frac{1}{2n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

The factor 1/2 simplifies the derivative when computing gradients.

## Part 3 Backpropagation

To improve the model, we compute the gradients of the loss with respect to the parameters.

This step is called backpropagation because the prediction error is propagated backward to update the parameters.

---

### Task 3.1 Gradient with Respect to \( w \)

The gradient of the loss with respect to \( w \) is:

$$
\frac{\partial L}{\partial w}
=
-\frac{1}{n}\sum_{i=1}^{n} x_i (y_i - \hat{y}_i)
$$

Compute

$$
\frac{\partial L}{\partial w}
$$

using the data and predictions from Part 2.

---

### Task 3.2 Gradient with Respect to \( b \)

The gradient with respect to \( b \) is:

$$
\frac{\partial L}{\partial b}
=
-\frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)
$$

Compute

$$
\frac{\partial L}{\partial b}
$$

### Solution

For \(w\):

$$
\frac{\partial L}{\partial w}
=
-\frac{1}{3} \left[1(0.1) + 2(0.6) + 3(0.0)\right]
=
-\frac{1}{3} (1.3)
=
-0.4333
$$

So,

$$
\frac{\partial L}{\partial w} \approx -0.4333
$$

For \(b\):

$$
\frac{\partial L}{\partial b}
=
-\frac{1}{3}(0.1 + 0.6 + 0.0)
=
-\frac{0.7}{3}
=
-0.2333
$$

So,

$$
\frac{\partial L}{\partial b} \approx -0.2333
$$


## Part 4 Gradient Descent

Once we know the gradients, we update the parameters using **gradient descent**.

Gradient descent moves parameters in the direction that reduces the loss.

The gradient points toward increasing loss, so we move in the opposite direction (hence we negate / minus).

The update rules are:

$$
w_{new} = w - \alpha \frac{\partial L}{\partial w}
$$

$$
b_{new} = b - \alpha \frac{\partial L}{\partial b}
$$

where

- $\alpha$ is the learning rate.

The learning rate controls the step size of the update.

For this exercise:

$$
\alpha = 0.1
$$

### Task 4.2 Interpretation

Answer the following questions:

1. Why do we subtract the gradient in the update rule?
2. What would happen if the learning rate $\alpha$ is too large?
3. What would happen if the learning rate $\alpha$ is too small?

### Solution

Using

$$
w_{new} = w - \alpha \frac{\partial L}{\partial w}
$$

$$
b_{new} = b - \alpha \frac{\partial L}{\partial b}
$$

with

- $(w = 0.5)$
- $(b = 2)$
- $(\alpha = 0.1)$
- $(\frac{\partial L}{\partial w} = -0.4333)$
- $(\frac{\partial L}{\partial b} = -0.2333)$

we get:

$$
w_{new} = 0.5 - 0.1(-0.4333) = 0.5433
$$

$$
b_{new} = 2 - 0.1(-0.2333) = 2.0233
$$

Therefore:

- $(w_{new} \approx 0.5433)$
- $(b_{new} \approx 2.0233)$

#### Task 4.2 Interpretation — Sample Answers

1. We subtract the gradient because the gradient points in the direction of increasing loss, and we want to move in the direction that reduces the loss.
2. If the learning rate \(\alpha\) is too large, the updates may overshoot the minimum and training may become unstable or diverge.
3. If the learning rate \(\alpha\) is too small, learning becomes very slow and may take many steps to improve the model.


### Concept Check

Explain how the following steps work together to train the model:

1. Forward pass
2. Loss calculation
3. Backpropagation
4. Gradient descent

### Solution

1. **Forward pass**: use the current values of \(w\) and \(b\) to compute predictions \(\hat{y}\).
2. **Loss calculation**: compare the predictions with the true values \(y\) to measure how wrong the model is.
3. **Backpropagation**: compute the gradients of the loss with respect to the parameters \(w\) and \(b\).
4. **Gradient descent**: update \(w\) and \(b\) using the gradients so that the loss becomes smaller.

Together, these four steps form one training cycle.
