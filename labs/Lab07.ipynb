## Part 1: Distance Based Measures

### Task 1.1: Distance Sensitivity Analysis  (Lecture 11 Slides 10 to 14)

Given:
- \(A = (1,2)\)
- \(B = (4,6)\)
- \(C = (1,10)\)

Compute L1, L2, and L∞ distances and determine which point is closer to A.


#### Solution

For A → B:

- L1 distance
  $[
  |4-1| + |6-2| = 3 + 4 = 7
  ]$

- L2 distance
  $[
  \sqrt{(4-1)^2 + (6-2)^2} = \sqrt{3^2 + 4^2} = 5
  ]$

- L∞ distance
  $[
  \max(|4-1|, |6-2|) = \max(3,4) = 4
  ]$

For A → C:

- L1 distance
  $[
  |1-1| + |10-2| = 0 + 8 = 8
  ]$

- L2 distance
  $[
  \sqrt{(1-1)^2 + (10-2)^2} = \sqrt{0 + 8^2} = 8
  ]$

- L∞ distance
  $[
  \max(|1-1|, |10-2|) = \max(0,8) = 8
  ]$

Conclusion: Point B is closer to A under all three metrics.

### Task 1.2: Scaling Trap (Lecture 11 Slide 18)

Dataset:

| Point | Age | Salary |
|---|---:|---:|
| P1 | 20 | 50,000 |
| P2 | 22 | 60,000 |
| P3 | 40 | 200,000 |

Does scaling affect nearest neighbour?


#### Solution

**Without scaling:**

For P1<->P2,

$[
\sqrt{(22-20)^2 + (60000-50000)^2}
=
\sqrt{2^2 + 10000^2}
\approx 10000
]$

For P1<->P3,

$[
\sqrt{(40-20)^2 + (200000-50000)^2}
=
\sqrt{20^2 + 150000^2}
\approx 150000
]$

So salary dominates the distance.

**Scaling:**

If we normalize both features to the range $([0,1])$:

- Age: 20 → 0, 22 → 0.10, 40 → 1.00  
- Salary: 50,000 → 0, 60,000 → 0.067, 200,000 → 1.00

Then:

For P1<->P2,
$[
\sqrt{(0.10-0)^2 + (0.067-0)^2}
\approx
\sqrt{0.01 + 0.004489}
\approx 0.120
]$

For P1<->P3,
$[
\sqrt{(1-0)^2 + (1-0)^2}
=
\sqrt{2}
\approx 1.414
]$

Conclusion: Scaling can change the geometry of the feature space, and therefore can change which points are considered nearest neighbours.

### Task 2.2: Compute Objective Function \(J\) (Lecture 11 Slide 19)

Using the final clusters from Q3, compute:
$J = \sum_{j=1}^{k}\sum_{x_i\in C_j} ||x_i - \mu_j||^2$


#### Solution

Using the centroids from Q3:

- $\mu_1 = (1.5,1)$
- $\mu_2 = (4.5,3.5)$

For **Cluster 1**:

Point $(1,1)$:
$(1-1.5)^2 + (1-1)^2 = 0.25 + 0 = 0.25$

Point $(2,1)$:
$(2-1.5)^2 + (1-1)^2 = 0.25 + 0 = 0.25$


For **Cluster 2**:

Point $(4,3)$:
$(4-4.5)^2 + (3-3.5)^2 = 0.25 + 0.25 = 0.5$

Point $(5,4)$:
$(5-4.5)^2 + (4-3.5)^2 = 0.25 + 0.25 = 0.5$

Therefore,
$J = 0.25 + 0.25 + 0.5 + 0.5 = 1.5$

**Conclusion:**  
The within-cluster sum of squares for this clustering is **1.5**.


### Task 2.3: Why does K-means fail on spiral data? (Lecture 11 Slides 36–37)

Explain why K-means often performs badly on spiral-shaped or otherwise irregular clusters.


#### Solution

K-means assigns each point to the nearest centroid, effectively splitting the space into regions where each region contains points closest to one centroid.

Because of this, K-means is best suited to clusters that are roughly:

- compact
- convex
- spherical around a centroid

Spiral-shaped clusters are non-convex.  
A point that belongs to one spiral arm may be closer in Euclidean distance to the centroid of another group, so K-means can split the structure incorrectly.

Conclusion:
K-means fails because its geometry is driven by distance to centroids, not by connected shape or density.


### Task 2.4: Does increasing \(k\) always improve clustering? (Lecture 11 Slides 29)

Discuss whether increasing the number of clusters always improves clustering quality.

#### Solution

Increasing $k$ always decreases or leaves unchanged the K-means objective:
$J = \sum_{j=1}^{k}\sum_{x_i\in C_j} ||x_i - \mu_j||^2$

So in that narrow optimization sense, increasing $k$ improves the fit.

However, this does not mean the clustering is more meaningful.

For example:

- If $k=n$, each point becomes its own cluster
- Then $J$=0
- But the clustering is trivial and useless

**Conclusion:**  
Increasing $k$ improves the objective value, but does not guarantee better real-world clustering quality.


### Task 2.5: Compute the silhouette score (Lecture 11 Slides 30-31)

For each point, compute:
$s(i)=\frac{b(i)-a(i)}{\max(a(i),b(i))}$

After computing $s(i)$ for each point, take the average of all $s(i)$ values to obtain the overall silhouette score.

Note: Silhouette score is computed using distances between data points, not distances to centroids.

#### Solution

Final clusters:
- $C_1 = \{(1,1), (2,1)\}$
- $C_2 = \{(4,3), (5,4)\}$

For each point:
$a(i) = \text{average distance to other points in same cluster}$

$b(i) = \text{average distance to points in nearest other cluster}$

$s(i)=\frac{b(i)-a(i)}{\max(a(i),b(i))}$

#### Point (1,1)

$a(i)=1$

$b(i)=\frac{\sqrt{13}+\sqrt{25}}{2}=\frac{3.61+5}{2}=4.3028$

$s(i)=\frac{4.305-1}{4.305}\approx 0.7676$

#### Point (2,1)

$a(i)=1$

$b(i)=\frac{\sqrt{8}+\sqrt{18}}{2}=\frac{2.83+4.24}{2}=3.5355$

$s(i)=\frac{3.535-1}{3.535}\approx 0.7172$

#### Point (4,3)

$a(i)=\sqrt{2}\approx 1.41$

$b(i)=\frac{\sqrt{13}+\sqrt{8}}{2}=\frac{3.61+2.83}{2}=3.2170$

$s(i)=\frac{3.22-1.41}{3.22}\approx 0.5604$

#### Point (5,4)

$a(i)=\sqrt{2}\approx 1.41$

$b(i)=\frac{5+\sqrt{18}}{2}=\frac{5+4.24}{2}=4.6213$

$s(i)=\frac{4.62-1.41}{4.62}\approx 0.6940$

#### Summary
| Point | a(i) | b(i) | s(i) |
|------|------:|------:|------:|
| (1,1) | 1.00 | 4.305 | 0.7676 |
| (2,1) | 1.00 | 3.535 | 0.7172 |
| (4,3) | 1.41 | 3.22  | 0.5604 |
| (5,4) | 1.41 | 4.62  | 0.6940 |

Average silhouette score:
$\bar{s}=\frac{0.7676+0.7172+0.5604+0.6940}{4}=0.6848$

$\boxed{\bar{s}\approx 0.685}$

### Task 3.2: What happens if \(\varepsilon\) is very large?  (Lecture 11 Slides 42-44)

Explain the effect of a very large \(\varepsilon\) in DBSCAN.


#### Solutions

If $\varepsilon$ becomes very large, then more points fall within each other’s neighbourhoods.

This causes:

- more points to satisfy the core condition
- fewer points to be labelled as noise
- clusters that were previously separate to merge together

In the extreme case, almost all points can become part of one giant cluster.

**Conclusion:**  
A very large $\varepsilon$ causes over-merging and may destroy meaningful cluster structure.
